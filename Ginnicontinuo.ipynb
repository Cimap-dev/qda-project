```python
# Funzione per applicare il gapping
def apply_gapping(data, track_id_col, gap=50):
    """
    Seleziona un valore ogni 'gap' per ogni track_id.
    Garantisce che almeno un valore venga selezionato per ogni track_id.
    """
    grouped = data.groupby(track_id_col)
    gapped_data = grouped.apply(lambda group: group.iloc[::gap] if len(group) > gap else group.iloc[[0, -1]].drop_duplicates()).reset_index(drop=True)
    return gapped_data

# Inizializza il dizionario per memorizzare i dataset gappati
gapped_datasets = {}

# Itera su ogni dataset con differenze
for file_path, data_diff in data_diff_datasets.items():
    print(f"\nElaborazione del file: {file_path}")
    
    # Verifica che la colonna 'track_id' esista
    if 'track_id' in data_diff.columns:
        # Applica il gapping
        gapped_data = apply_gapping(data_diff, track_id_col='track_id', gap=50)
        
        # Crea un nuovo dataset gappato
        gapped_datasets[file_path] = gapped_data
        print(f"Dataset gappato creato per: {file_path}")
    else:
        print(f"Colonna 'track_id' non trovata nel file {file_path}")
```

```python
for file_name, gapped_data in gapped_datasets.items():
    print(f"\nProcessing ARIMA for file: {file_name}")
    
    # Check if the 'temp' column exists in the gapped dataset
    if 'temp' in gapped_data.columns:
        # Apply ARIMA model
        model = qda.ARIMA(gapped_data['temp'], order=(0, 1, 1), add_constant=False)
        qda.ARIMAsummary(model)
        residuals = model.resid.dropna()  # Drop NaN values from residuals

        # Anderson-Darling test on residuals
        result_resid = anderson(residuals, dist='norm')
        print(f"Anderson-Darling statistic (residuals): {result_resid.statistic:.4f}")
        for sl, cv in zip(result_resid.significance_level, result_resid.critical_values):
            print(f"Significance Level {sl}%: Critical Value {cv:.4f}")

        # Durbin-Watson test on residuals
        dw = durbin_watson(residuals)
        print(f"Durbin-Watson statistic: {dw}")
    else:
        print(f"Column 'temp' not found in the dataset for file: {file_name}")
```
