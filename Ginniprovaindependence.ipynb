```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.stats import anderson
from statsmodels.stats.stattools import durbin_watson
import qdatoolkit as qda

# === Config ===
FOLDER_PATH = './data'
GAP_SIZE = 50

# === Functions ===
def load_csv_files(folder):
    return [os.path.join(folder, f) for f in os.listdir(folder) if f.endswith('.csv') and os.path.getsize(os.path.join(folder, f)) > 0]

def compute_temp_diff(data):
    if 'temp' not in data.columns:
        return None
    data['temp_diff'] = data['temp'].diff()
    return data.drop(columns='temp').rename(columns={'temp_diff': 'temp'}).dropna()

def apply_gapping(data, group_col='track_id', gap=GAP_SIZE):
    if group_col not in data.columns:
        return None
    return data.groupby(group_col).apply(
        lambda g: g.iloc[::gap] if len(g) > gap else g.iloc[[0, -1]].drop_duplicates()
    ).reset_index(drop=True)

def plot_and_test(series, title):
    series.plot(title=f"Prima differenza di 'temp' - {title}")
    plt.show()
    sm.qqplot(series, line='s')
    plt.title(f"QQ Plot - {title}")
    plt.show()
    result = anderson(series, dist='norm')
    print(f"Anderson-Darling statistic: {result.statistic:.4f}")
    for sl, cv in zip(result.significance_level, result.critical_values):
        print(f"Significance Level {sl}%: Critical Value {cv:.4f}")

def run_arima_analysis(series, title):
    model = qda.ARIMA(series, order=(0, 1, 1), add_constant=False)
    qda.ARIMAsummary(model)
    residuals = model.resid.dropna()
    print(f"\nResidual analysis for: {title}")
    plot_and_test(residuals, f"Residuals - {title}")
    print(f"Durbin-Watson statistic: {durbin_watson(residuals)}")

# === Main Execution ===
file_paths = load_csv_files(FOLDER_PATH)

for path in file_paths:
    print(f"\nAnalisi del file: {path}")
    df = pd.read_csv(path)
    diff_df = compute_temp_diff(df)
    if diff_df is None:
        print("Colonna 'temp' non trovata.")
        continue
    plot_and_test(diff_df['temp'], path)

    gapped_df = apply_gapping(diff_df)
    if gapped_df is None:
        print("Colonna 'track_id' non trovata.")
        continue
    run_arima_analysis(gapped_df['temp'], path)
```

```python
durbin_results = []

for path in file_paths:
    print(f"\nProcessing file: {path}")
    df = pd.read_csv(path)
    diff_df = compute_temp_diff(df)
    if diff_df is None:
        print("Column 'temp' not found.")
        continue

    gapped_df = apply_gapping(diff_df)
    if gapped_df is None:
        print("Column 'track_id' not found.")
        continue

    residuals = qda.ARIMA(gapped_df['temp'], order=(0, 1, 1), add_constant=False).resid.dropna()
    dw_stat = durbin_watson(residuals)
    durbin_results.append({'file': path, 'durbin_watson_stat': dw_stat})

print(durbin_results)
```
