```python
import pandas as pd

# Replace 'your_file.csv' with the path to your CSV file
data = pd.read_csv('./data/layer_324.csv')

# Display the first few rows of the dataframe
data.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import qdatoolkit as qda

# Perform the runs test for independence on the 'temp' column
result = qda.Assumptions(data['temp']).independence()

# Display the result
print(result)
```

```python
# Combine two variables into a single sequence
combined_sequence = np.concatenate([data['temp'], data['t_rel']])

# Perform the runs test for independence
result_combined = qda.Assumptions(combined_sequence).independence()

# Display the result
print(result_combined)
```
