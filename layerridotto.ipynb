```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import qdatoolkit as qda

layer226 = pd.read_csv('./data/layer_226.csv')

```

```python
plt.figure(figsize=(8, 6))
plt.scatter(layer226['x'], layer226['y'], s=1, label='Layer 226')
plt.title('Layer 226')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#calcolo ymin
# Filtra i dati per track_id pari e maggiore o uguale a 85
even_track_ids = layer226[(layer226['track_id'] >= 84)]

# Raggruppa per track_id e ottieni il primo valore di y
first_y_values = even_track_ids.groupby('track_id')['y'].min()
print(first_y_values)
# Trova il valore minimo
max_value = first_y_values.max()

# Stampa i risultati

print(f"Maximum value: {max_value}")



```

```python

#calcolo ymax
# Filtra i dati per track_id pari e maggiore o uguale a 85
last_track_ids = layer226[(layer226['track_id'] <= 84)]

# Raggruppa per track_id e ottieni il primo valore di y
last_y_values = last_track_ids.groupby('track_id')['y'].max()
print(last_y_values)
# Trova il valore minimo
min_value = last_y_values.min()

# Stampa i risultati

print(f"Minimum value: {min_value}")


```

```python

#calcolo xmax
# Filtra i dati per track_id pari e maggiore o uguale a 85
last_track_ids_x = layer226[(layer226['track_id'] >= 84)& (layer226['track_id'] != 172)]

# Raggruppa per track_id e ottieni il primo valore di y
last_x_values = last_track_ids_x.groupby('track_id')['x'].max()
last_x_values = last_x_values[last_x_values > 37]
print(last_x_values)
# Trova il valore minimo
min_value = last_x_values.min()

# Stampa i risultati

print(f"Minimum value: {min_value}")
```

```python

#calcolo xmin
# Filtra i dati per track_id pari e maggiore o uguale a 85
first_track_ids_x = layer226[(layer226['track_id'] <= 84)]

# Raggruppa per track_id e ottieni il primo valore di y
first_x_values = first_track_ids_x.groupby('track_id')['x'].min()

print(first_x_values)
# Trova il valore minimo
max_value = first_x_values.max()

# Stampa i risultati

print(f"Maximum value: {max_value}")

```

```python
ymin = -66.522
ymax = -54.62
xmin = 26.155
xmax = 38.06

range_x = xmax - xmin
range_y = ymax - ymin
print(f"Range of x: {range_x}")
print(f"Range of y: {range_y}")
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(filtered_dataset_226['x'], filtered_dataset_226['y'], s=1)
plt.title('Layer 226')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
