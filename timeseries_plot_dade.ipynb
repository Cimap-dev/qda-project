```python
import pandas as pd
import glob
import os


def read_all_csv_files(folder_path):
    # Get all CSV file paths in the folder
    csv_files = glob.glob(os.path.join(folder_path, "*.csv"))
    
    # Read each CSV file and store them in a list of DataFrames
    dataframes = [pd.read_csv(file) for file in csv_files]
    return dataframes
    

folder_path = "./data"
all_data = read_all_csv_files(folder_path)
```

```python
stacked_data = pd.concat(all_data, ignore_index=True)
```

```python
import matplotlib.pyplot as plt

def plot_temp_time_series(data, n=100):
    sampled_data = data.iloc[::n]  # Sample every n observations
    plt.figure(figsize=(12, 6))
    plt.scatter(sampled_data['t'], sampled_data['temp'], label='Temperature', alpha=0.7)
    sampled_data['t'] = sampled_data['t'] / 1e6  # Convert microseconds to seconds
    plt.xlabel('t (seconds)')
    plt.legend()
    plt.grid()
    plt.show()

plot_temp_time_series(stacked_data, n=1)
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(all_data[0]['t'], all_data[0]['temp'], label='Temperature', s=1)
plt.xlabel('Time (t)')
plt.ylabel('Temperature (temp)')
plt.title('Temperature Time Series')
plt.legend()
plt.grid()
plt.show()
```
